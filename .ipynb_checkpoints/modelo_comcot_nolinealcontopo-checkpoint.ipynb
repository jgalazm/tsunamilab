# COMCOT lineal con batimetría

* $h:$ coordenada vertical del fondo marino, $>0$ en el agua $<0$ en tierra
* $\eta:$ coordenada vertical de la superficie libre, $>0$ implica el agua está sobre el nivel medio del mar, $\eta=-h$ en tierra.
* $H=h+\eta$ altura de la columna del agua, $H=0$ en tierra
* el $0$ vertical está en el nivel medio del agua WML

```python
def simulation(x,h,eta,p,dx,cfl,nt):
    dp1 = np.zeros_like(x)
    dp2 = np.zeros_like(x)
    zdepth = np.zeros_like(eta)
    zdepth[0,:] = eta[0,:]+h

    for n in range(nt-1):
        ## Mas Loop
        for i in range(1, nx-1):
            zzz = eta[n,i] - rx*(p[n,i]-p[n,i-1])
            DD = zzz + h[i]
            zdepth[n+1,i] = DD
            eta[n+1,i] = zzz    


        # Calculate total water depth and total water depth a discharge point
        for i in range(nx):
            ip1 = min(nx-1,i+1)
            dp1[i] = 0.25*(zdepth[n+1,i] + zdepth[n,i] + zdepth[n+1,ip1] + zdepth[n, ip1])
            dp2[i] = 0.5*(zdepth[n+1,i]+zdepth[n+1,ip1])    

        # open boundary
        i = 0
        if h[i]>gx:
            cc = np.sqrt(9.81*h[i])
            uh = 0.0 #de acuerdo al codigo fortran
            uu = np.sqrt(uh**2+p[n,i]**2)
            zz = uu/cc
            if p[n,i]>0:
                zz = -zz
            eta[n+1,i] = zz
            zdepth[n+1,i] = zz+h[i]#esto no esta en el codigo original

        i = nx-1
        if h[i]>gx:
            cc = np.sqrt(9.81*h[i])
            uh = 0.0 #de acuerdo al codigo fortran
            uu = np.sqrt(uh**2+p[n,i-1]**2)
            zz = uu/cc
            if p[n,i-1]<0:
                zz = -zz
            eta[n+1,i] = zz     
            zdepth[n+1,i] = zz+h[i]##esto no esta en el codigo original

        # momentum loop

        for i in range(0,nx-1):
            noflux = False
            #moving boundary
            ip1 = min(nx-1,i+1)
            c1 = (zdepth[n+1,i]>0.0 and zdepth[n+1,ip1]>0.0)
            c2 = (zdepth[n+1,i]>0.0 and zdepth[n+1,ip1]<=0.0 \
                 and h[ip1]+eta[n+1,i]>0)
            c3 = (zdepth[n+1,i]<=0.0 and zdepth[n+1,ip1]>0.0 \
                 and h[i]+eta[n+1,ip1]>0)
            if (zdepth[n+1,i]>0.0 and zdepth[n+1,ip1]>0.0):
                dd = dp2[i]
                df = dp1[i]
            elif (zdepth[n+1,i]>0.0 and zdepth[n+1,ip1]<=0.0 \
                 and h[ip1]+eta[n+1,i]>0):
                dd = h[ip1] + eta[n+1,i]
                df = dd
            elif (zdepth[n+1,i]<=0.0 and zdepth[n+1,ip1]>0.0 \
                 and h[i]+eta[n+1,ip1]>0):
                dd = h[i] + eta[n+1,ip1]
                df = dd
            else:
                p[n+1,i] = 0.0
                noflux = True
            if not noflux:
                p[n+1,i] = p[n,i] - 9.81*rx*dd*(eta[n+1,i+1]-eta[n+1,i])
    return h,eta,p
```

## Caso de prueba #1

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nt = 1000
nx = 500
x = np.linspace(-100,100,nx)

h = np.zeros_like(x)
h = np.where(np.abs(x)<=30,1.0,2)
eta = np.zeros((nt,nx))
eta[0,:] = 0.5*np.exp(-(x+60)**2/50) ##gaussian bell
#eta[0,:] = np.where(np.abs(x-50)<20, 0.04, 0.0) #square
# eta[0,:] = 8*1/8.*np.where(np.abs(x+30)<20,np.exp(-1./(1-((np.abs(x)+30)/20)**2)),0.0) #mollifier bell
# eta = np.where(x<10,-h,eta)
# eta = np.where(x>90,-h,eta)
eta[0,:] = np.where(h>=0,eta[0,:],-h)

p = np.zeros((nt,nx))

plt.plot(x,eta[0,:])
plt.plot(x,-h)
plt.ylim(-5,3.0)

cfl = 0.9
dx = np.diff(x)[0]
dt = cfl*dx/np.sqrt(np.max(h)*9.81)
rx = dt/dx
gx = 1e-5
grx = 9.81*dt/dx
```

```python
h,eta,p = simulation(x,h,eta,p,dx,cfl,nt)
```

```python
# create a simple animation
fig = plt.figure(figsize=(12,4))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
line1, = ax1.plot([], [], lw=2)
line2, = ax2.plot([], [], lw=2)
ax1.plot(x,-h,color='g')

ax1.set_xlim(-100,100)
ax1.set_ylim(-5,3)
ax2.set_xlim(-100,100)
ax2.set_ylim(-1,1)

def init():
    line1.set_data([], [])
    line2.set_data([],[])
    return line1,line2

def animate(i):
    line1.set_data(x, eta[i*4,:])
    line2.set_data(x, p[i*4,:])
    return line2,

animation.FuncAnimation(fig, animate, init_func=init,
                        frames=250, interval=20, blit=True)
```
