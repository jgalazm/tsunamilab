# Guardar batimetria como archivo jpg

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from clawpack.geoclaw import topotools
import clawpack.clawutil.data as clawutildata
topo_fname = 'etopo10min120W60W60S0S.asc'
url = 'http://www.geoclaw.org/topo/etopo/' + topo_fname
clawutildata.get_remote_file(url, output_dir='.', file_name=topo_fname,
        verbose=True)

data = np.loadtxt('etopo10min120W60W60S0S.asc',skiprows=6)
topo = np.reshape(data,(361,361))
lon = np.arange(-120.,-120+361*0.1666666667,0.1666666667)
lat = np.arange(-60.,-60+361*0.1666666667,0.1666666667)
```

```python
import utm
xy = [utm.from_latlon(lat[i],lon[i],force_zone_number=18)[:2] for i in range(lat.shape[0])]
xy = np.array(xy)
```

revisar que los arreglos sean crecientes

```python
plt.subplot(131)
plt.plot(xy[:,0])
plt.subplot(132)
plt.plot(xy[:,1])
plt.subplot(133)
plt.plot(xy[:-1,1])
```

2 problemas: En x no son crecientes, y en $y$ explota en el último. Hay que interpolar

La topografía original

```python
x,y = np.meshgrid(xy[:-1,0],xy[:-1,1])
z = np.flipud(topo[:-1,:-1])
plt.pcolormesh(x,y,z)
```

Preparo la malla de los datos interpolados

```python
xi = np.linspace(x[0,0],x[0,-1],x.shape[0])
yi = np.linspace(y[0,0],y[-1,0],y.shape[0])
xi,yi = np.meshgrid(xi,yi)
```

Interpolo para generar una malla

```python
zi = plt.mlab.griddata(x.ravel(), y.ravel(), z.ravel(),xi,yi,interp='linear')
```

```python
plt.pcolormesh(xi,yi,zi)
```

```python
plt.pcolormesh(xi,yi,zi.mask)
plt.colorbar()
```

Es un arreglo enmascarado, pero ningun valor está en la mascara

## Truncar la batimetria

```python
zt = np.trunc( (zi-zi.min())/(zi.max()-zi.min())*255. )
err = zt*(zi.max()-zi.min())/255.+zi.min()-zi
errrel = np.where( 1-np.isclose(zi,0.0), np.abs(err)/zi, 0.0)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.pcolormesh(zt)
plt.colorbar()
plt.title('bati truncada')
plt.axis('off')

plt.subplot(132)
plt.pcolormesh(x,y,err,cmap=plt.cm.hot)
plt.colorbar()
plt.title('error absoluto\n(cuantificacion)')
plt.axis('off')

plt.subplot(133)
plt.pcolormesh(x,y,np.log(errrel),vmin=-10.,vmax=10.)
plt.colorbar()
plt.title('log - error relativo\n(cuantificacion)')
plt.axis('off')

plt.tight_layout()
```

## Crear la batimetría y el archivo de info batimétrica

```python
import scipy
```

```python
from scipy.misc import toimage
```

```python
im = scipy.misc.toimage(np.flipud(z))
im.save('bati2.jpg')
```

Propiedades de la batimetría

```python
f = open('bati2.txt','w')
f.write('ny   :%i\n'%z.shape[0])
f.write('nx   :%i\n'%z.shape[1])
f.write('xmin :%.3f\n'%x.min())
f.write('xmax :%.3f\n'%x.max())
f.write('ymin :%.3f\n'%y.min())
f.write('ymax :%.3f\n'%y.max())
f.write('zmin :%.3f\n'%z.min())
f.write('zmax :%.3f\n'%z.max())
f.close()
```
